```python
# loc, iloc 이용한 데이터 접근 방법
# loc 단일 선택 -> row 기준
df.loc[0]
# 다중 선택
df.loc[[0,2,5]]
# 슬라이싱 : 어디서 어디까지 잘라서 가져와라
df.loc[0:5]
df2=df.loc[0:5:2]
# 논리 인덱싱 : 중요. True/False 말함. 논리값 갯수는 row 개수랑 일치해야함. flase로 쓰면 해당 row 출력 안함.
df2.loc[[True, True, True]]

# row와 column을 같이 선택 : 단일인지 다중인지 슬라이싱인지 논리인지는 응용하기에 따라 다 사용가능
df.loc[0,'찾으신금액']
df.loc[0:5,['찾으신금액','맡기신금액',	'거래후 잔액'	]]

# 특정 조건을 만족하는 행만 선택하고 싶을 때
# 예) 찾으신 금액이 10만원 이상인 행만 선택.

# 판다스에서는 for문과 if 문으로 따로 안해도 아래와 같이 하면 자동으로 논리 인덱싱을 해줌.
df['찾으신금액']>100000
# 아래와 같이 하면 True인 row의 내용이 출력
df.loc[df['찾으신금액']>100000]
# 이거 문법 잘 알아두기! 문법때매 헤메것네
df.loc[(df['찾으신금액']>100000) & (df['적요']=='통신요금')]
```

```python
import pandas as pd
from tables import index
# 1. students파일을 읽어서 데이터 프레임으로 만든 후 아래를 해결해주세요.
stu = pd.read_excel('/content/Github/students.xlsx')
# 2. 국어가 80점 이상인 학생의 이름과 생년월일
stu.loc[stu['국어']>=80,['이름','생년월일']]

# 3. 수학이 80점 미만인 학생의 영어와 국어 점수
stu.loc[stu['수학']<80,['이름','영어','국어']]

# 4. 국어와 수학이 90점 이상인 학생의 이름, 수학, 국어 점수
stu.loc[(stu['국어']>=90)&(stu['수학']>=90),['이름','수학','국어']]

# 5. 모든 과목이 60점 미만인 학생의 번호, 이름, 생년월일
stu.loc[(stu['수학']<60)&(stu['국어']<60)&(stu['영어']<60),['번호','이름','생년월일']]

# 6. 대전 지역 학생들의 이름, 국어, 영어, 수학 점수
stu.loc[stu['지역명']=='대전']

# 문제1. 학생 데이터를 국어 성적으로 내림차순 정렬한 뒤 위에서 11개, 아래에서 7개 삭제
stu_sorted=stu.sort_values(by='국어', ascending=False)

stu_remove1=stu_sorted.iloc[11:]

stu_remove2=stu_remove1[:-7]

stu_remove2
```

```python
#  iloc로 데이터 접근
# 인덱스명을 숫자가 아닌 a,b,c라든가 야식1,야식2 등으로 변경한 경우. loc[0]을 하면 0이라는 인덱스가 없기 때문에 가져오지 않는다.
# 이때 iloc를 이용하면(iloc[0]) 수정 전의 0번 인덱스로 인식. 가져옴.
```
